```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
quar_name = 'quarantine.csv'
folder_path = r'..\covid-19'
quar_path = os.path.join(folder_path, quar_name)
```

```python
df_qrt = pd.read_csv(quar_path)
df_qrt['Fecha'] = pd.to_datetime(df_qrt['Fecha'])
df_qrt.sort_values(by='Fecha', inplace=True)
df_qrt
```

```python
in_cnd = (df_qrt['Estado'] == 'in') & (df_qrt['Nivel'] == 'full')
dt_cnd =  df_qrt['Fecha'] > pd.to_datetime('2020-04-12')
df_qrt_in = df_qrt[in_cnd & dt_cnd]

gdf_in = df_qrt_in.groupby('Fecha')

for grp, df in gdf_in:
    display(df)
```
