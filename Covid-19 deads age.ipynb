```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = 'FallecidosEtario_T.csv'
folder_path = r'..\Datos-COVID19\output\producto10'
file_path = os.path.join(folder_path, file_name)
```

```python
df = pd.read_csv(file_path)
display(df)
```

```python
x_tim = df['Grupo de edad'].values
y_39 = df['<=39'].values
y_49 = df['40-49'].values
y_59 = df['50-59'].values
y_69 = df['60-69'].values
y_79 = df['70-79'].values
y_89 = df['80-89'].values
y_90 = df['>=90'].values
plt.title('Chile')
plt.plot(x_tim, y_39, label='<=39')
plt.scatter(x_tim, y_39)
plt.plot(x_tim, y_49, label='40-49')
plt.scatter(x_tim, y_49)
plt.plot(x_tim, y_59, label='50-59')
plt.scatter(x_tim, y_59)
plt.plot(x_tim, y_69, label='60-69')
plt.scatter(x_tim, y_69)
plt.plot(x_tim, y_79, label='70-79')
plt.scatter(x_tim, y_79)
plt.plot(x_tim, y_89, label='80-89')
plt.scatter(x_tim, y_89)
plt.plot(x_tim, y_90, label='>=90')
plt.scatter(x_tim, y_90)
plt.legend(loc='upper left')
plt.xlabel('days')
plt.ylabel('Cummulative Deads')
plt.xticks(rotation=45)
#plt.yscale('log')
plt.grid()
plt.show()
```

```python
lst39 = [] ; lst49 = [] 
lst59 = [] ; lst69 = [] 
lst79 = [] ; lst89 = []
lst90 = []

for ind in df.index:
    ntot = float(df.iloc[ind][1:].values.sum())
    lst39.append(100.0 * df['<=39'][ind]  / float(ntot))
    lst49.append(100.0 * df['40-49'][ind] / ntot)
    lst59.append(100.0 * df['50-59'][ind] / ntot)
    lst69.append(100.0 * df['60-69'][ind] / ntot)
    lst79.append(100.0 * df['70-79'][ind] / ntot)
    lst89.append(100.0 * df['80-89'][ind] / ntot)
    lst90.append(100.0 * df['>=90'][ind]  / ntot)

df['%, <=39']  = lst39
df['%, 40-49'] = lst49
df['%, 50-59'] = lst59
df['%, 60-69'] = lst69
df['%, 70-79'] = lst79
df['%, 80-89'] = lst89
df['%, >=90']  = lst90
df
```

```python
x_tim = df['Grupo de edad'].values
yp_39 = df['%, <=39'].values
yp_49 = df['%, 40-49'].values
yp_59 = df['%, 50-59'].values
yp_69 = df['%, 60-69'].values
yp_79 = df['%, 70-79'].values
yp_89 = df['%, 80-89'].values
yp_90 = df['%, >=90'].values
plt.title('Chile')
plt.plot(x_tim, yp_39, label='<=39')
plt.scatter(x_tim, yp_39)
plt.plot(x_tim, yp_49, label='40-49')
plt.scatter(x_tim, yp_49)
plt.plot(x_tim, yp_59, label='50-59')
plt.scatter(x_tim, yp_59)
plt.plot(x_tim, yp_69, label='60-69')
plt.scatter(x_tim, yp_69)
plt.plot(x_tim, yp_79, label='70-79')
plt.scatter(x_tim, yp_79)
plt.plot(x_tim, yp_89, label='80-89')
plt.scatter(x_tim, yp_89)
plt.plot(x_tim, yp_90, label='>=90')
plt.scatter(x_tim, yp_90)
plt.legend(loc='upper left')
plt.xlabel('days')
plt.ylabel('Daily percentage of deads by age')
plt.xticks(rotation=45)
#plt.yscale('log')
plt.grid()
plt.show()
```

```python
fig, axs = plt.subplots(2, sharex=True, figsize=(8,8))
fig.suptitle('Cummulative chilean deads by age',  y=0.99, fontsize = 22)

axs[0].plot(x_tim, y_39, label='<=39')
axs[0].scatter(x_tim, y_39)
axs[0].plot(x_tim, y_49, label='40-49')
axs[0].scatter(x_tim, y_49)
axs[0].plot(x_tim, y_59, label='50-59')
axs[0].scatter(x_tim, y_59)
axs[0].plot(x_tim, y_69, label='60-69')
axs[0].scatter(x_tim, y_69)
axs[0].plot(x_tim, y_79, label='70-79')
axs[0].scatter(x_tim, y_79)
axs[0].plot(x_tim, y_89, label='80-89')
axs[0].scatter(x_tim, y_89)
axs[0].plot(x_tim, y_90, label='>=90')
axs[0].scatter(x_tim, y_90)
axs[0].set_ylabel('Cummulative Deads', fontsize = 15)
axs[0].legend(loc='upper left', fontsize = 12)
axs[0].tick_params('y', labelsize=12)
axs[0].grid()

axs[1].plot(x_tim, yp_39, label='<=39')
axs[1].scatter(x_tim, yp_39)
axs[1].plot(x_tim, yp_49, label='40-49')
axs[1].scatter(x_tim, yp_49)
axs[1].plot(x_tim, yp_59, label='50-59')
axs[1].scatter(x_tim, yp_59)
axs[1].plot(x_tim, yp_69, label='60-69')
axs[1].scatter(x_tim, yp_69)
axs[1].plot(x_tim, yp_79, label='70-79')
axs[1].scatter(x_tim, yp_79)
axs[1].plot(x_tim, yp_89, label='80-89')
axs[1].scatter(x_tim, yp_89)
axs[1].plot(x_tim, yp_90, label='>=90')
axs[1].scatter(x_tim, yp_90)
axs[1].set_ylabel('Percentage of Deads', fontsize = 15)
axs[1].set_xlabel('Days', fontsize = 15)
axs[1].legend(loc='upper left', fontsize = 12)
axs[1].grid()
axs[1].tick_params('x', labelrotation=45, labelsize=12)
axs[1].tick_params('y', labelsize=12)

fig.tight_layout()
plt.subplots_adjust(top=0.94)

image_name = 'Covid19ChileDeadsAges.jpg'
folder_path = r'C:\Users\HN566KT\Downloads'
image_path = os.path.join(folder_path, image_name)
plt.savefig(image_path)
```
